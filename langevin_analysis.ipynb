```python
from scipy.stats import gamma,expon
xg=np.linspace(0.5,1.5,100)
hist_D=plt.hist(trace['D'],200,normed=1)
mean_D=trace['D'].mean()
std_D=trace['D'].std()
print('D mean: ',mean_D,'std: ',std_D)
scale=std_D**2/mean_D
alpha=mean_D/scale
print('D alpha: ',alpha,'scale: ',scale)
gy=gamma.pdf(xg,alpha,scale=scale)
plt.plot(xg,gy)
plt.plot(xg,expon.pdf(xg,scale=10))
```

```python
hist_A=plt.hist(trace['A'],200,normed=1)
xA=np.linspace(0,4,100)
mean_A=trace['A'].mean()
std_A=trace['A'].std()
print('mean: ',mean_A,'std: ',std_A)
scaleA=std_A**2/mean_A
alphaA=mean_A/scaleA
print('alpha: ',alphaA,'scale: ',scaleA)
gA=gamma.pdf(xA,alphaA,scale=scaleA)
plt.plot(xA,gA)
plt.plot(xg,expon.pdf(xA,scale=5))
```
